```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

**Loading The Image**

```python
img = cv2.imread('/content/sample_data/Sunflower.jpg',1)
cv2_imshow(img)
```

**Converting an image into grey scale image**

```python
img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_grey)
```

**Inverting The Image**

```python
img_invert = cv2.bitwise_not(img_grey)
cv2_imshow(img_invert)
```

**Smoothing The Image**

```python
img_smoothing = cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
cv2_imshow(img_smoothing)
```

**Obtaining Final Sketch**

```python
def dodgeV2(x,y):
  return cv2.divide(x,255-y,scale =256)
```

**Final Output as Pencil Sketch**

```python
final_img = dodgeV2(img_grey,img_smoothing)
cv2_imshow(final_img)
```
